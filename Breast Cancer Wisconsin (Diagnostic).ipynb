```python
# Drop the 'id' and 'Unnamed: 32' columns
data = data.drop(['id', 'Unnamed: 32'], axis=1)

# Encode the 'diagnosis' column into numerical values
le = LabelEncoder()
data['diagnosis'] = le.fit_transform(data['diagnosis'])

# Separate features and target variable
X = data.drop('diagnosis', axis=1)
y = data['diagnosis']

# Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_scaled = pd.DataFrame(X_scaled, columns=X.columns)
```

```python
# Replace infinite values with NaN
X_scaled.replace([np.inf, -np.inf], np.nan, inplace=True)

# Check for any remaining NaN values and fill them
X_scaled.fillna(X_scaled.mean(), inplace=True)
```

```python
# Select the top k features
selector = SelectKBest(f_classif, k=10)
X_reduced = selector.fit_transform(X_scaled, y)

# Get the selected feature names
selected_features = X.columns[selector.get_support()]
```

```python
# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_reduced, y, test_size=0.2, random_state=42)
```

```python
# Initialize the Logistic Regression model
log_reg = LogisticRegression()

# Fit the model to the training data
log_reg.fit(X_train, y_train)
```

```python
# Predict on the test data
y_pred_log_reg = log_reg.predict(X_test)
```

```python
from sklearn.svm import SVC

# Initialize the Support Vector Classifier
svc = SVC(kernel='linear')  # You can change the kernel to 'rbf', 'poly', etc.

# Fit the model to the training data
svc.fit(X_train, y_train)

# Predict on the test data
y_pred_svc = svc.predict(X_test)
```

```python
from sklearn.ensemble import RandomForestClassifier

# Initialize the Random Forest Classifier
rf = RandomForestClassifier(n_estimators=100, random_state=42)  # You can adjust the number of trees

# Fit the model to the training data
rf.fit(X_train, y_train)

# Predict on the test data
y_pred_rf = rf.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, classification_report

# Evaluate the performance of Logistic Regression
accuracy_log_reg = accuracy_score(y_test, y_pred_log_reg)
print(f'Logistic Regression Accuracy: {accuracy_log_reg}')
print(classification_report(y_test, y_pred_log_reg))

# Evaluate the performance of SVM
accuracy_svc = accuracy_score(y_test, y_pred_svc)
print(f'SVM Accuracy: {accuracy_svc}')
print(classification_report(y_test, y_pred_svc))

# Evaluate the performance of Random Forest
accuracy_rf = accuracy_score(y_test, y_pred_rf)
print(f'Random Forest Accuracy: {accuracy_rf}')
print(classification_report(y_test, y_pred_rf))
```
